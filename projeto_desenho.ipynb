#### PyAutoGUI:
- PyAutoGUI é uma API que permite seus scripts Python controlar o mouse e o teclado para automatizar  
as interações com outros aplicativos, funciona com Windows, macOS e Linux e em Python 2 e 3.
- Ferramenta legal para crianção automatizada de alguns desenhos, esse pequeno projeto é uma simples  
demonstração das utilidades de algumas funções.
- Você pode consultar a documentação no link https://pyautogui.readthedocs.io/en/latest/

- Os valores passados dentro das funções __[pyautogui]__ são respectivamente os  
eixos __[X]__, __[Y]__ e __[duração]__.
